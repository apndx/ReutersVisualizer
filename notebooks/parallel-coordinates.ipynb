## Parallel coordinates

I will try out the data with parallel coordinates, and maybe bar charts as well

```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas
import json
import re
```

```python
# Opening JSON file
f = open('data/used_topics_sorted.json',)
  
# returns JSON object as 
# a dictionary
used_topics_sorted = json.load(f)
```

```python
used_topics_sorted
```

```python
filename = "data/country_geo_topic_counts.gpkg"
reut_country_geo_topic = geopandas.read_file(filename)
reut_country_geo_topic.set_index('country', inplace=True)
reut_country_geo_topic['topiccounts'] = reut_country_geo_topic['topiccounts'].apply(eval)
reut_country_geo_topic
```

```python
reuters_with_topic_cols = pd.read_csv('csv/reuters-with-topic-cols-bin.csv', delimiter=';')
reuters_with_topic_cols
```

```python
reuters_with_topic_cols = reuters_with_topic_cols.set_index('id')
reuters_with_topic_cols['desc'] = reuters_with_topic_cols['desc'].apply(eval)
reuters_with_topic_cols['codes'] = reuters_with_topic_cols['codes'].apply(eval)
reuters_with_topic_cols.fillna('', inplace=True)
print(reuters_with_topic_cols.info())
reuters_with_topic_cols
```

```python
reut_topic_col_geom = pd.merge(reuters_with_topic_cols, reut_country_geo_topic, how='left', on='country')
reut_topic_col_geom 
```

```python
# df with one row per country

countries_topics_count_each = pd.read_csv('data/countries_topics_count_each.csv', delimiter=';')
countries_topics_count_each
```

```python
country_topic_col_geom = pd.merge(countries_topics_count_each, reut_country_geo_topic, how='left', on='country')
country_topic_col_geom 
```

```python
first_10_countries = country_topic_col_geom.head(10)
```

```python
# all countries, all topics
fig_1 = px.parallel_coordinates(country_topic_col_geom, dimensions=['country', 'count', *used_topics_sorted], color='count', color_continuous_scale=px.colors.sequential.Inferno)

fig_1.show()
```

```python
# first 10 countries
fig_2 = px.parallel_coordinates(first_10_countries, dimensions=['country', 'count', *used_topics_sorted], color='count', color_continuous_scale=px.colors.sequential.Inferno)

fig_2.show()
```

```python
top_ten_topics = ['CORPORATE/INDUSTRIAL',
 'GOVERNMENT/SOCIAL',
 'MARKETS',
 'PERFORMANCE',
 'ECONOMICS',
 'ACCOUNTS/EARNINGS',
 'COMMODITY MARKETS',
 'COMMENT/FORECASTS',
 'MONEY MARKETS',
 'DOMESTIC POLITICS']

# first 10 topics, first 10 countries
fig_3 = px.parallel_coordinates(first_10_countries, dimensions=['country', 'count', *top_ten_topics], color='count', color_continuous_scale=px.colors.sequential.Inferno)

fig_3.show()
```

```python
# first 10 topics, first 10 countries
fig_4 = px.parallel_categories(first_10_countries, dimensions=['country', 'count', *top_ten_topics], color='count', color_continuous_scale=px.colors.sequential.Inferno)

fig_4.show()
```

```python
# first 5 topics, first 5 countries
top_five_topics = ['CORPORATE/INDUSTRIAL',
 'GOVERNMENT/SOCIAL',
 'MARKETS',
 'PERFORMANCE',
 'ECONOMICS']

# first 10 topics, first 10 countries
fig_5 = px.parallel_categories(country_topic_col_geom.head(5), dimensions=['country', 'count', *top_five_topics], color='count', color_continuous_scale=px.colors.sequential.Inferno)

fig_5.show()
```

### Bar chart with plotly

Let's try plotting document amounts in different countries with bar charts

```python
filename = "data/country_geo_topic_counts.gpkg"
reut_country_geo_topic = geopandas.read_file(filename)
#reut_country_geo_topic.set_index('country', inplace=True)
reut_country_geo_topic['topiccounts'] = reut_country_geo_topic['topiccounts'].apply(eval)
reut_country_geo_topic
```

```python
# lets do a version where the dict is string

filename = "data/country_geo_topic_counts.gpkg"
reut_country_geo_topic_strings = geopandas.read_file(filename)
#reut_country_geo_topic.set_index('country', inplace=True)
#reut_country_geo_topic['topiccounts'] = reut_country_geo_topic['topiccounts'].apply(eval)

reut_country_geo_topic_strings
```

```python

reut_country_geo_topic_strings['topiccounts2']= reut_country_geo_topic_strings['topiccounts'].apply(lambda x: x.replace(',', ',\n'))
reut_country_geo_topic_strings
```

```python
sum_of_all_docs = reut_country_geo_topic['count'].sum()
sum_of_all_docs
reut_country_geo_topic['percentage'] = round((reut_country_geo_topic['count']/sum_of_all_docs)*100, 3)
reut_country_geo_topic
```

```python
import plotly.express as px

# this is the one

fig_6 = px.bar(reut_country_geo_topic, x='country', y='count',
             hover_data=['count', 'percentage'], color='count',
             labels={'percentage':'Percentage (%)', 'count':'Amount of documents', 'country': 'Country'}, height=800)
fig_6.show()
```

```python
import plotly.graph_objects as go


# Use the hovertext kw argument for hover text
fig_7 = go.Figure(data=[go.Bar(x=reut_country_geo_topic_strings['country'], y=reut_country_geo_topic_strings['count'],
            hovertext=reut_country_geo_topic_strings['topiccounts2'])])
# Customize aspect
fig_7.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig_7.update_layout(title_text='Countries and topics')
fig_7.show()
```
